```python
import matplotlib.pyplot as plt
```

```python
movies  = [
    "the dark knight",
    "the hurt locker",
    "the kings speech",
    "the artist",
    "argo"
]
years = [2008,2009,2010,2011,2012]
revenue = [1005,170,427,133,232] # in million dollors
```

```python
plt.bar(years,revenue,color = "green")
plt.title("Revenue in each year for oscar movies")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
for i in range(len(years)):
    plt.text(years[i],revenue[i]+5,str(revenue[i]),ha = "center")
plt.ylim(0, max(revenue)+200)
```
