```python
import matplotlib.pyplot as plt
```

```python
oscar_movies  = [
    "the dark knight",
    "the hurt locker",
    "the kings speech",
    "the artist",
    "argo"
]
years = [2008,2009,2010,2011,2012]
oscar_revenue = [1005,170,427,133,232] # in million dollors

non_oscar_movies  = [
    "slumdog millionair",
    "avatar",
    "Inception",
    "Lincoln"
]
non_oscar_revenue = [378,2788,829,185,275]
```

```python
import numpy as np
x = np.arange(len(years))
width = 0.4
plt.bar(x-width/2,oscar_revenue , width,label = "Oscar Movies")
plt.bar(x+width/2,non_oscar_revenue,width,label = "Non Oscar Movies")
plt.title("Oscar vs Non Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue(in $M)")
plt.legend()
plt.xticks(x,years)
```
