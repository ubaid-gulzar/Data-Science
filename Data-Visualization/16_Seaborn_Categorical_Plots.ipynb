```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
print(tips.head(20))
```

```python
#barplot
sns.catplot(
    data = tips,
    kind = "bar",
    x = "day",
    y = "tip",
    hue = "sex"
)
```

```python
#barplot
sns.catplot(
    data = tips,
    kind = "box",
    x = "day",
    y = "tip",
)
```
