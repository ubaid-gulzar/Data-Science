```python
#Real data
#oscur winner movies
oscar_movies  = [
    "the dark knight",
    "the hurt locker",
    "the kings speech",
    "the artist",
    "argo"
]
years = [2008,2009,2010,2011,2012]
oscar_revenue = [1005,170,427,133,232] # in million dollors

#plot which year has what revenue
plt.plot(years,oscar_revenue)
plt.title("Oscar movies Revenue in each year")
plt.xlabel("Years")
plt.ylabel("Revenue(in $M)")
```

```python
#oscur winner movies
non_oscar_movies  = [
    "slumdog millionair",
    "avatar",
    "Inception",
    "Lincoln"
]
non_oscar_revenue = [378,2788,829,185,275]
plt.plot(years,oscar_revenue,".--r",label = "Oscar Movies")
plt.plot(years,non_oscar_revenue, "o-g",label = "Non Oscar Movies")
plt.title("Oscur Movies Revenue vs Non Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue(in $M)")
plt.legend()
```

```python
plt.plot(years,oscar_revenue,label = "Oscar Movies")
plt.plot(years,non_oscar_revenue ,label = "Non Oscar Movies")
plt.title("Oscur Movies Revenue vs Non Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue(in $M)")
plt.legend()

#other methods
plt.style.available
plt.style.use('dark_background')
plt.grid()
plt.show()
plt.savefig("01_plot.png")
```
