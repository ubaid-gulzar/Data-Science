```python
import matplotlib.pyplot as plt
```

```python
#Multiple data sets
legit_transactions = [
    2.99,5.49,8.99,12.50,14.99,19.99,23.45,29.99,34.99,39.50,
    45.00,49.99,55.25,60.00,75.99,89.99,120.50,150.00,199.00,
    249.99,300.75,450.00,600.00,850.00,1200.00 
]

fraud_transactions = [
    50,100,150,200,300,500,500,750,1000,1000,
    1200,1500,1500,2000,2500,3000,3000
]
```

```python
plt.hist(legit_transactions,color = "green" ,label = "Legit Transactions",bins = 10,edgecolor  = "black",alpha = 0.5)
plt.hist(fraud_transactions,color = "red" ,label = "Fraud Transactions",bins = 10,edgecolor  = "black",alpha = 0.5)
plt.title("Transaction Amount Disturbution - Legit vs Fruad")
plt.xlabel("Amount (in $)")
plt.xlabel("Frequency")
plt.legend()
```
