```python
import matplotlib.pyplot as plt
```

```python
expenses = ["Salaries","Rent","Markiting","R&D","Miscellaneous"]
amount = [500,150,120,100,50]
```

```python
exp = [0,0,0,0.1,0]
plt.pie(amount,labels = expenses, autopct = "%1.1f%%", wedgeprops = {
    "edgecolor" : "black",
    "linewidth": 2,
    "linestyle" : "--",
    "alpha": 0.9,
}, explode = exp
       )
```
