# Compare temp. of cities over a week using subplots and loops

```python
import matplotlib.pyplot as plt
```

```python
days =["Mon", "Tue" ,"Wed" ,"Thu", "Fri"]
cities =["New York", "London" ,"Delhi" ,"Tokyo"]
temp = [
    [22,23,24,24,25],   #new york
    [18,19,17,20,21],   #london
    [30,32,31,33,34],   #delhi
    [25,26,24,27,11]    #Tokyo
]
```

```python
fig,ax = plt.subplots(2,2)
count = 0
for i in range(2):
    for j in range(2):
        ax[i][j].plot(days,temp[count])
        ax[i][j].set_title(cities[count])
        ax[i][j].set_ylim(15,35)
        ax[i][j].grid()
        count += 1
fig.suptitle("Temperature in cities over the week")
fig.supxlabel("Days")
fig.supylabel("Temperature")
fig.tight_layout()
```
