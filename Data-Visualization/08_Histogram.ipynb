```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
np.random.seed(0)
score = np.random.normal(70,10,1000)  #normal disturbution
plt.hist(score,color = "brown" , edgecolor = "black",bins = 5)
plt.xlabel("Score")
plt.ylabel("Frequency of students")
```

```python
#cutum made bins
np.random.seed(0)
score = np.random.normal(70,10,1000)  #normal disturbution
bins = [0,30,70,90,100]
plt.hist(score,color = "brown" , edgecolor = "black",bins = bins)
plt.xlabel("Score")
plt.ylabel("Frequency of students")
```
