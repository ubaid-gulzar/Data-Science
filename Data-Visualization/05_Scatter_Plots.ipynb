```python
import matplotlib.pyplot as plt
```

```python
people = ["person A ","person B","person C ","person D","person E ",
          "person F","person G ","person H","person I ","person J"]
age = [22,25,30,35,40,45,50,55,60,65]
bp = [110,115,120,122,125,130,135,123,145,150]

plt.scatter(age,bp)
plt.title("Age vs Blood Presure")
plt.xlabel("Age")
plt.ylabel("Blood Presure")
plt.grid()
```

```python
#customization
colors = ["green" if x < 135 else  "red" for x in bp]
plt.scatter(age,bp,color = colors,s =20,alpha = 0.9)
plt.title("Age vs Blood Presure", color = "green")
plt.xlabel("Age")
plt.ylabel("Blood Presure")
plt.grid()
```

```python
#colormap
#customization
plt.scatter(age,bp,cmap = "plasma",c = bp)
plt.title("Age vs Blood Presure", color = "green")
plt.xlabel("Age")
plt.ylabel("Blood Presure")
plt.colorbar()
```

```python
#annotation
plt.scatter(age,bp,cmap = "rainbow",c = bp)
plt.title("Age vs Blood Presure", color = "green")
plt.xlabel("Age")
plt.ylabel("Blood Presure")
for i in range(len(people)):
    plt.annotate(people[i],xy = (age[i],bp[i]), xytext = (age[i],bp[i]+1))
plt.xlim(min(age),max(age)+10)
plt.ylim(min(bp),max(bp)+5)
plt.grid()
```
