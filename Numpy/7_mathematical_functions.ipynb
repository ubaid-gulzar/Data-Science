```python
import numpy as np
```

```python
#Aggrigation
arr = np.array([1,2,3,4,5])
print("sum of array:",np.sum(arr))
print("product of array:",np.prod(arr))
print("minimum value in array:",np.min(arr))
print("maximum value in array:",np.max(arr))
print("index of min value:",np.argmin(arr))
print("index of max value:",np.argmax(arr))
print("index of min value:",np.argmin(arr))
print("mean of array(values in array):",np.mean(arr))
print("standard devation:",np.std(arr))
print("median of array",np.median(arr))
print("variance is:",np.var(arr))
```

```python
#power functions
arr = np.array([1,2,3,4,5])
print("square of every element in array:",np.square(arr))
print("square root of every element in array:",np.sqrt(arr))
print("power 3 of every element in array:",np.power(arr,3))
```

```python
#log and exponential
arr = np.array([1,2,3,4,5])
print("Natural log:",np.log(arr))
print("Base 2 log:",np.log2(arr))
print("Base 10 log:",np.log10(arr))
print("exponent :",np.exp(arr))
```

```python
#rounding
arr = np.array([1.2,2.6,3,4,5])
print("roundof 3.14 =",np.round(3.55))
print("roundof of array =",np.round(arr))
print("ceil 3.14 =",np.ceil(3.14))
print("floor of 5.9:",np.floor(5.9))

arr = np.array([1,2,5,3,7,4,5,2,1])
print("unique =",np.unique(arr))
print("sorted",np.sort(arr))

arr = np.array([1,-2,-6,3,5])
print("absolute positive array is:",np.abs(arr))
```
