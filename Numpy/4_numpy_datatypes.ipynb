```python
import numpy as np
```

```python
#data types(integer)
arr= np.array([1,2,3,4,5])
print(arr.dtype)
arr = arr.astype(np.int32)
print(arr.dtype)
arr = arr.astype(np.int16)
print(arr.dtype)
arr = arr.astype(np.int8)
print(arr.dtype)
```

```python
#data types(floating)
arr= np.float32(np.array([1.2,2,3,4,5]))
print(arr,arr.dtype)
```

```python
#data types(boolean)
arr= np.array([True,False,False,True])
print(arr,arr.dtype)
```

```python
#data types(strings)
arr= np.array(["ubaid","gulzar"])
print(arr,arr.dtype)
```

```python
#data types(complex)
arr= np.array([5+6j,7+2j])
print(arr,arr.dtype,arr.shape)
```
