```python
import pandas as pd
```

```python
df = pd.read_csv("08_Data_cleaning.csv")
df
```

```python
#apply
df2 = df.copy()
def tax(x):
    if(x>50000):        
        return x + x*0.2
    else:
        return x + x*0.1
df2["income"] = df2["income"].apply(tax)
df2
```

```python
#map
gender_map ={"male": "M","female":"F","unknown":"U"}
df2["gender"] = df2["gender"].map(gender_map)
df2
```

```python
#assign
df2.assign(new_income = df2["income"] *1.1)
```

```python
#replace(old,new): replaces old values with some new values (repace USA with US)
df2["country"].replace("usa","United States")
```
