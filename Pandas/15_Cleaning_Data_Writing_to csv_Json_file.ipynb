```python
import pandas as pd 
```

```python
df = pd.read_csv("08_Data_cleaning.csv")
df
```

```python
df2 = df.copy()
df2
```

```python
#drop duplicated rows
df2 = df2.drop_duplicates()
df2
```

```python
#Fill income and age  NAN values with 0
df2[["income","age"]] = df2[["income","age"]].fillna(0)
#fill gender and name unknown values with Male and NAN with male
df2["gender"] = df2["gender"].replace("unknown","male").fillna("male")
#FILL COUNTRY NAN Values with UNknown
df2["country"] = df2["country"].fillna("unknown")
#fill name NAN values with Unknown
df2["name"] = df2["name"].fillna("unknown")
df2
```

```python
#sort the dataframe on the basis of income
df2 = df2.sort_values("income")
df2 = df2.reset_index(drop = True)
df2
```

```python
#store the data into new csv file
df2.to_csv("15_Sorted Data_of_08_csv_file.csv")
```

```python
#store the data into new json file
df2.to_json("15_Sorted Data_of_08_csv_file.json")
```
