```python
import pandas as pd
import numpy as np
```

```python
#creating dataFrame from object
info = {
    "names" : ["ubaid","kamran","sahil"],
    "CGPA" : [9.5,9.9,5.5],
    "university" : "IUST"
}
df = pd.DataFrame(info)
print(df)
print(df.index)
print(df.columns)
```

```python
#creating DataFrame from list of lists
df = pd.DataFrame([["Adam",23],["ubaid",21],["Bob",20]],columns = ["name","age"])
print(df)
```

```python
#creating DataFrame from numpy array
arr = np.array([[1,2,3],[4,5,6]])
df = pd.DataFrame(arr,columns = ["a","b","c"])
print(df)
```
