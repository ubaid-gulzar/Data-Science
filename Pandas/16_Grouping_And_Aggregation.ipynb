```python
import pandas as pd
```

```python
df = pd.read_csv("08_Data_cleaning.csv")
df
```

```python
#GROUPIN: ON THE BASIS OF COUNTRY
print(df.groupby("country")["income"].mean())
print(df.groupby("country")["income"].min())
```

```python
df.groupby("gender")["age"].mean()
```

```python
df.groupby("country")["income"].agg(["mean","median","min","max"])
```

```python
df4 = df.groupby("country")["income"].agg(["mean","median","min","max"])
df4
```

```python
df4 = df.groupby("country")["income"].agg(min_salary = "min",max_salary = "max",avg_salary ="mean")
df4
```

```python
df.groupby("country").agg({"income":"mean","age":"mean"})
```

```python
df.groupby("country").agg(avg_income = ("income","mean"),min_age = ("age","min"))
```
