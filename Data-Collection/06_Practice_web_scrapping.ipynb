```python
#From Quotes extract all quotes & authors from all pages with the life tag
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://quotes.toscrape.com/page/1/"
page_count = 1
while True:
    URL = f"https://quotes.toscrape.com/page/{page_count}/"
    res = requests.get(URL)
    soup = BeautifulSoup(res.text, "lxml")
    quote = soup.select("div.quote")
    if not quote:
        print("no valid pages anymore total pages =",page_count)
        break
    with open(f"06_scrapped_data_quotes{page_count}.html","w" ,encoding ="utf-8") as f:
        f.write(res.text)
        print(f"sucessuflly downloaded data from page{page_count}")
    page_count += 1
```

```python
#Extract useful information
life_quotes = []
for i in range(1,page_count):
    with open(f"06_scrapped_data_quotes{i}.html","r",encoding = "utf-8") as f:
        html_content = f.read()
    soup = BeautifulSoup(html_content,"lxml")
    all_quotes = soup.select("div.quote")
    for q in all_quotes:
        all_tags = []
        for tag in q.select("a.tag"):
            all_tags.append(tag.get_text())
        if "life" in all_tags:
            text = q.select_one("span.text").get_text()
            author = q.select_one("small.author").get_text()
            life_quotes.append([text,author])
print(life_quotes)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(life_quotes, columns = ["quotes","author"])
df
```

```python
df.to_csv("All_Life_Quotes.csv")
```
