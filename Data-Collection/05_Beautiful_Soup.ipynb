```python
from bs4 import BeautifulSoup
```

```python
with open("03_scrapped_data.html", "r") as f:
    html_content = f.read()
```

```python
soup = BeautifulSoup(html_content,"lxml")
```

```python
h1 = soup.find("h1")
print(h1)
```

```python
h3_all = soup.find_all("h3")

for h3 in h3_all:
    name = h3.get_text(strip = True)
```

```python
for h3 in h3_all:
    print(h3.find_parent()["class"])
```

```python
#finds next element after the h1
h1.find_next("h3")
```

```python
#finds element before the h1
h1.find_previous()
```

```python
# find the population of every country
population = soup.find_all("span", class_ = "country-population")
for pop in population:
    print(pop.get_text(strip = True))
```

```python
for h3 in h3_all:
    print(h3.find_next("div").select("span.country-population")[0].get_text())
```
